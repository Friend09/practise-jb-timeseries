## heat_map_yearly

```python
ds.head()
```

```python
years = pd.DataFrame()
groups = ds.groupby(pd.Grouper(freq="YE"))
```

```python
for name, group in groups:
    years[name.year] = group.values

years
```

```python
years.T
```

```python
# years in y axis, days on x-axis
plt.matshow(years.T, interpolation=None, aspect="auto")
```

## stacked_line_plot

```python
ds
```

```python
groups = ds.groupby(pd.Grouper(freq="YE"))
years = pd.DataFrame()
```

```python
for name, group in groups:
    years[name.year] = group.values

years
```

```python
years.plot(subplots=True, legend=False)
plt.show()
```
